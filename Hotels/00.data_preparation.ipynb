```python
prompt = f"""
I will provide you with a hotel description in English.

You must translate it to Korean.Then, put information into json format with the following schema

<JSONSCHEMA>
- "hotel": The name of the hotel
- "address": Korean Address
- "description": The summary description in Korean
- "facilities": The facilities in Korean, separated by comma
- "attractions": The attractions information near the hotel in Korean, format of name(distance), separated by comma
- "airport": The nearest airport in Korean
- "summary": Summary of the hotel including all the information about location, description, facilities, attractions and the nearest airport
</JSONSCHEMA>

Here's an instrunction.

<INSTRUCTION>
- The answer must be in strict json format with no markdown or other comment
- Ignore html tag.
- Be as detail as possible.
- If there's no information provided for description, say "정보가 업데이트 되지 않았습니다"
- Translate "Distances are displayed to the nearest 0.1 mile and kilometer" into "주요 관광지"
- Ignore HTML tags.
<INSTRUCTION>

Here's hotel information

<DOC>
{text_to_translate}
</DOC>
"""

prompt_config = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 10000,
    "temperature": 0.5,
    "top_k": 350,
    "top_p": 0.999,
    "messages": [
        {
            "role": "user",
            "content": [{"type": "text", "text": prompt}],
        }
    ],
}

body = json.dumps(prompt_config)
body
```

```python
# Anthropic Claude를 사용하여 응답 생성하기
session = boto3.Session()
retry_config = Config(
    region_name=os.environ.get("us-east-1", None),
    retries={
        "max_attempts": 10,
        "mode": "standard",
    },
)

bedrock = session.client(
    service_name="bedrock-runtime", config=retry_config
)  # Bedrock 클라이언트를 생성합니다.

response = bedrock.invoke_model(
    body=body,
    modelId="anthropic.claude-3-sonnet-20240229-v1:0",
    contentType="application/json",
    accept="application/json",
)

response_body = json.loads(response.get("body").read())
results = response_body.get("content")[0].get("text")
print(results)
```

```python
def call_claude(hotel_name):
    text_to_translate = f"""
        <hotel>
            <name>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["HotelName"].values[0]}</name>
            <city>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["cityCode"].values[0]}</city>
            <address>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["Address"].values[0]}</address>
            <description>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["Description"].values[0]}</description>
            <attraction>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["Attractions"].values[0]}</attraction>
            <rating>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["HotelRating"].values[0]}</rating>
            <map>{kr_hotels[kr_hotels["HotelName"] == hotel_name]["Map"].values[0]}</map>
        </hotel>
    """

    prompt = f"""
        I will provide you with a hotel description in English.

        You must translate it to Korean.Then, put information into json format with the following schema

        <JSONSCHEMA>
        - "hotel": The name of the hotel, must be in Korean
        - "city": The city code of the hotel, must be in Korean
        - "rating": The rating of the hotel in number
            - OneStar: 1
            - TwoStar: 2
            - ThreeStar: 3
            - FourStar: 4
            - FiveStar: 5
        - "address": Korean Address
        - "description": The summary description in Korean
        - "facilities": The facilities in Korean, separated by comma
        - "attractions": The attractions information near the hotel in Korean, format of name(distance), separated by comma
        - "airport": The nearest airport in Korean
        - "summary": Summary of the hotel including all the information about location, description, facilities, attractions and the nearest airport
        - "map": The location of the hotel, format of longitude,latitude delimited by '|'
        </JSONSCHEMA>

        Here's an instrunction.

        <INSTRUCTION>
        - The answer must be in strict json format with no markdown or other comment
        - Ignore html tag.
        - Be as detail as possible.
        - If there's no information provided for description, say "정보가 업데이트 되지 않았습니다"
        - Translate "Distances are displayed to the nearest 0.1 mile and kilometer" into "주요 관광지"
        - Ignore HTML tags.
        <INSTRUCTION>

        Here's hotel information

        <DOC>
        {text_to_translate}
        </DOC>
    """

    prompt_config = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 10000,
        "temperature": 0.5,
        "top_k": 350,
        "top_p": 0.999,
        "messages": [
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": prompt},
                ],
            }
        ],
    }

    body = json.dumps(prompt_config)

    # It's possible that it failed to invoke model, then we just skip this hotel
    try:
        response = bedrock.invoke_model(
            body=body,
            modelId="anthropic.claude-3-sonnet-20240229-v1:0",
            contentType="application/json",
            accept="application/json",
        )
    except Exception as e:
        print("bedrock.invoke_model failed")
        return None

    response_body = json.loads(response.get("body").read())
    results = response_body.get("content")[0].get("text")

    # Check if results is valid json format
    try:
        answer = json.loads(results)
        return answer

    except json.decoder.JSONDecodeError:
        return None
```

```python
# Iterate kr_hotels dataframe, get hotel name and call call_claude function
# and check the status bar using tqdm package
from tqdm.notebook import tqdm
import time
import csv
import os


def inference_on_file(start_index=0):
    schema = [
        "name",
        "city",
        "rating",
        "address",
        "description",
        "facilities",
        "attractions",
        "airport",
        "summary",
        "map",
    ]

    kr_hotels_korean = pd.DataFrame(columns=schema)

    output_file = "kr_hotels_korean.csv"
    # if output_file doesn't exist, create it
    if not os.path.exists(output_file):
        with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=schema)
            writer.writeheader()

    for idx in tqdm(range(start_index, len(kr_hotels))):
        hotel = kr_hotels.iloc[idx]
        hotel_name = hotel.HotelName
        answer = call_claude(hotel_name)
        if answer is not None:
            new_record_df = pd.DataFrame.from_dict(answer, orient="index").T
            new_record_df.to_csv(output_file, mode="a", header=False, index=False, encoding="utf-8")

            kr_hotels_korean = pd.concat(
                [kr_hotels_korean, new_record_df],
                ignore_index=True,
            )

            print(f"Processed {idx}th hotel, {hotel_name}")
        else:
            continue

    print(f"finished processing {idx} hotels")
```

```python
# And I want to get all the items that latitude is greater than longitude.
# If latitude is greater than longitude, they should switch positions.
def switch_lat_long(row):
    lat, long = map(float, row.split(","))
    if lat > long:
        return f"{long},{lat}"
    return row


hotels_raw_df["map"] = hotels_raw_df["map"].apply(switch_lat_long)
```
